```python
selected_features = ['station_avg_temp_c','reanalysis_dew_point_temp_k','reanalysis_specific_humidity_g_per_kg','total_cases','station_min_temp_c']
train_data_sj = train_data_sj[selected_features]
train_data_iq = train_data_iq[selected_features]
train_data_sj.describe()
```

```python
train_data_iq.describe()
```

```python
train_sj_subtrain = train_data_sj.head(800)
train_sj_subtest = train_data_sj.tail(train_data_sj.shape[0] - 800)

train_iq_subtrain = train_data_iq.head(400)
train_iq_subtest = train_data_iq.tail(train_data_iq.shape[0] - 400)
```

```python
grid = 10 ** np.arange(-8, -3, dtype=np.float64)
print(grid)
```

```python
#using lasso regression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error

def lasso_regressor(train_data,validation_data, alpha):
    features = ['station_avg_temp_c','reanalysis_dew_point_temp_k','reanalysis_specific_humidity_g_per_kg','station_min_temp_c']
    
    #training dataset features
    train_features = train_data[features]
    #training dataset lables
    train_labels = train_data[["total_cases"]]
    
    #validation dataset features
    valid_features = validation_data[features]
    #training dataset lables
    valid_labels = validation_data[["total_cases"]]
    
    #Fit the model
    lassoreg = Lasso(alpha=alpha,normalize=True, max_iter=1e5)
    lassoreg.fit(train_features,train_labels)
    y_pred = lassoreg.predict(valid_features)
    
    mae = mean_absolute_error(valid_labels,y_pred)
    
    print("Mean Absolute Error for :",alpha," ",mae)
    
    return lassoreg
```

```python

pred_dataset = deng_features_test_df[['station_avg_temp_c','reanalysis_dew_point_temp_k','reanalysis_specific_humidity_g_per_kg','station_min_temp_c']]

pred_dataset_sj = pred_dataset.loc['sj']
pred_dataset_sj = pred_dataset_sj.fillna(pred_dataset.mean())

pred_dataset_iq = pred_dataset.loc['iq']
pred_dataset_iq = pred_dataset_iq.fillna(pred_dataset.mean())
```

```python
model_sj = lasso_regressor(train_sj_subtrain,train_sj_subtest,0.000000000000000001)
predictions_sj = model_sj.predict(pred_dataset_sj).astype(int)
```

```python
model_iq = lasso_regressor(train_iq_subtrain,train_iq_subtest,0.00000000000000000000000000000000000000001)
predictions_iq = model_iq.predict(pred_dataset_iq).astype(int)
```

```python
submission = pd.read_csv("./data/submission_format.csv",
                         index_col=[0, 1, 2])

submission.total_cases = np.concatenate([predictions_sj, predictions_iq])
submission.to_csv("./data/submission2.csv")
```
